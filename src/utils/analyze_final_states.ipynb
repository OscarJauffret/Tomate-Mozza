```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

with open("../../final_states.csv") as f:
    lines = f.readlines()

lines = [line.removeprefix('[').removesuffix(']\n').split(",") for line in lines]
lines = [[float(x) for x in line] for line in lines]
lines = np.array(lines)

# Convert to DataFrame
df = pd.DataFrame(lines, columns=["section_rel_x", "section_rel_y", "next_turn" ,"velocity", "acceleration", "relative_yaw", "second_edge_length", "second_turn", "third_edge_length", "third_turn"])


# Plotting
plt.figure(figsize=(15, 10))
plt.subplot(2, 5, 1)
plt.hist(df["section_rel_x"], bins=50)
plt.title("section_rel_x")
plt.subplot(2, 5, 2)
plt.hist(df["section_rel_y"], bins=50)
plt.title("section_rel_y")
plt.subplot(2, 5, 3)
plt.hist(df["next_turn"], bins=50)
plt.title("next_turn")
plt.subplot(2, 5, 4)
plt.hist(df["velocity"], bins=50)
plt.title("velocity")
plt.subplot(2, 5, 5)
plt.hist(df["acceleration"], bins=50)
plt.title("acceleration")
plt.subplot(2, 5, 6)
plt.hist(df["relative_yaw"], bins=50)
plt.title("relative_yaw")
plt.subplot(2, 5, 7)
plt.hist(df["second_edge_length"], bins=50)
plt.title("second_edge_length")
plt.subplot(2, 5, 8)
plt.hist(df["second_turn"], bins=50)
plt.title("second_turn")
plt.subplot(2, 5, 9)
plt.hist(df["third_edge_length"], bins=50)
plt.title("third_edge_length")
plt.subplot(2, 5, 10)
plt.hist(df["third_turn"], bins=50)
plt.title("third_turn")
plt.tight_layout()
plt.show()
```
